```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('./dataset/sonar.csv', header=None)

'''
print(df.info())
print(df.head())
'''

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=seed)

# 성능 차이 설명.
#테스트 셋과 학습 셋의 비율이 30:70 이었을때
#Test Accuracy: 0.8571, Study Accuracy: 1.0000
#테스트 셋과 학습 셋의 비율이 40:60 이었을때
#Test Accuracy: 0.7857, Study Accuracy: 1.0000
#테스트 셋의 비율을 높이니까 테스트 셋의 정확도는 떨어졌으나, 학습 셋의 정확도는 같다.

model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=130, batch_size=5)

print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
print("\n Study Accuracy: %.4f" % (model.evaluate(X_train, Y_train)[1]))
```
